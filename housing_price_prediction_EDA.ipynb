**Missing values**

```python

dataset.isnull().sum()
```

```python
##we will take the features having NAN values
feature_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]

##we will calculate feature name and percentage of missing values
for features in feature_na:
   print(features,np.round(dataset[features].isnull().mean(),4),' % of missing values')
```

###**Since there are many missing values we need to find the relation ship between missing values and salesprice**

```python

for feature in feature_na:
   data=dataset.copy()
##lets make a variable that indicates 1 if the observation was missing else 0 
   data[feature]=np.where(data[feature].isnull(),1,0)

#lets calculate the mean salesprice for missing and not missing
   data.groupby(feature)['SalePrice'].median().plot.bar() 
   plt.title(feature)
   plt.show()
```

Here with the relation between missing values and the dependent variables is clearly visible.so we need to replace the nan values with something meaningful which we will do in the feature engineering section.

from the above dataset some of the features like ID is not required.

**Numerical Variables**

```python
## we will calculate the no of numerical variables in features
numerical_features=[features for features in dataset.columns if dataset[features].dtypes !='O']
print(len(numerical_features))
```

```python
dataset[numerical_features].head()
```

###**Temporal Variables(Eg: Datetime Variables)**

From the dataset we have 4 year variables.we have extract information from the datetime variables like no of years or no of days.one example in this specific scenario can be difference in years between the year the house was built and the year the house was sold .we will be performing this analysis in the feature engineering which is the next video.

```python
## we calculated the year features
year_feature=[features for features in numerical_features if  'Year'in features or 'Yr' in features ]
year_feature
```

```python
for feature in year_feature:
  print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
```

```python
data=dataset.copy()
for features in year_feature:
  if features !='YrSold':
     data[features]=data['YrSold']-data[features]
     plt.scatter(data[features],data['SalePrice'])
     plt.xlabel(features)
     plt.ylabel('SalePrice')
     plt.show()
```

```python
##numerical features are divided into two: continous features and discrete features
discrete_features=[features for features in numerical_features if len(dataset[features].unique())<25 and features not in year_feature and feature!='Id']
discrete_features
```

```python
print('Discrete variables_count=',len(discrete_features))
```

```python
##we want to analyse the relationship between the discrete feature and dependent feature
for features in discrete_features:
  dataset.groupby(features)['SalePrice'].median().plot.bar()
  plt.xlabel(features)
  plt.ylabel('Saleprice')
  plt.show()
```

**Continous Features**

```python
##we initialise continous features into list
continous_features=[features for features in numerical_features if features not in discrete_features+year_feature+['Id']]
len(continous_features)
```

```python
for features in continous_features:
  plt.hist(dataset[features],bins=50)
  plt.xlabel(features)
  plt.ylabel('count')
  plt.show()
```

**Exploratory Data Analysis part2**

```python
##we will use logarithmic Transformtion
data=dataset.copy()
data['Saleprice']=np.log(data['SalePrice'])
for features in continous_features:
  if 0 not in data[features].unique():
    data[features]=np.log(data[features])
    plt.scatter(data[features],data['SalePrice'])
    plt.xlabel(features)
    plt.ylabel('Saleprice')
    plt.show()
  else:
    pass
  
```

##**Outliers**

```python
data=dataset.copy()
data['SalePrice']=np.log(data['SalePrice'])
for features in continous_features:
  if 0 not in data[features].unique():
    data[features]=np.log(data[features])
    data.boxplot(column=features)
    plt.show()
  else:
    pass
```
